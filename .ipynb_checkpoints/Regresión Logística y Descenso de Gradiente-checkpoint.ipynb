## Función de costes

La Función de Costes es una función que podemos utilizar para evaluar si nuestro algoritmo mapea la estimación objetivo, o si nuestro algoritmo realiza correctamente problemas de optimización.


En primer lugar, debemos definir la Función Sigmoide: 

```python
def logistic_function(x):    
    return 1/ (1 + np.exp(-x))
```

Definimos la Funcion de Costes: 

```python
def compute_cost(theta, x, y):
    m = len(y)
    y_pred = logistic_function(np.dot(x , theta))
    error = (y * np.log(y_pred)) + ((1 - y) * np.log(1 - y_pred))
    cost = -1 / m * sum(error)
    gradient = 1 / m * np.dot(x.transpose(), (y_pred - y))
    return cost[0] , gradient
```

```python
mean_puntuacion = np.mean(puntuacion, axis=0)
std_puntuacion = np.std(puntuacion, axis=0)
puntuaciones = (puntuacion - mean_puntuacion) / std_puntuacion
rows = puntuaciones.shape[0]
cols = puntuaciones.shape[1]

X = np.append(np.ones((rows, 1)), puntuaciones, axis=1) #include intercept
y = resultados.reshape(rows, 1)

theta_init = np.zeros((cols + 1, 1))
cost, gradient = compute_cost(theta_init, X, y)

print("Cost at initialization", cost)
print("Gradient at initialization:", gradient)
```

El 0.693147180559946 es el con los parámetros iniciales.

## Descenso del Gradiente


El Descenso de Gradiente es un algoritmo de optimización para encontrar un mínimo local de una función diferenciable. El Descenso de Gradiente se utiliza para encontrar los valores de los parámetros (coeficientes) de una función que minimizan una Función de Coste en la medida de lo posible.

```python
def gradient_descent(x, y, theta, alpha, iterations):
    costs = []
    for i in range(iterations):
        cost, gradient = compute_cost(theta, x, y)
        theta -= (alpha * gradient)
        costs.append(cost)
    return theta, costs
```

Representamos el Descenso de Gradiente con 200 interacciones: 

```python
theta, costs = gradient_descent(X, y, theta_init, 1, 200)
```

```python
plt.plot(costs)
plt.xlabel("Iteraciones")
plt.ylabel("$J(\Theta)$")
plt.title("Valor de la Funcion de Coste sobre las Iteraciones del Valor de Descenso del Gradiente con 200 interacciones")
```

Podemos observar que con 200 interacciones, el valor de la Función de Costes se acerca a un mínimo. 
Vamos a aumentar las interacciones para comprobar como reacciona la Funcion de Costes.

Representamos el Descenso de Gradiente con 400 interacciones:

```python
theta, costs = gradient_descent(X, y, theta_init, 1, 400)
```

```python
plt.plot(costs)
plt.xlabel("Iteraciones")
plt.ylabel("$J(\Theta)$")
plt.title("Valor de la Funcion de Coste sobre las Iteraciones del Valor de Descenso del Gradiente con 200 interacciones")
```

Comprobamos que con 400 interacciones, el valor de la Función de Costes es menor que con 200 interacciones. 

```python
theta, costs = gradient_descent(X, y, theta_init, 2, 100)
```

```python
plt.plot(costs)
plt.xlabel("Iteraciones")
plt.ylabel("$J(\Theta)$")
plt.title("Valor de la Funcion de Coste sobre las Iteraciones del Valor de Descenso del Gradiente")
```

También cabe la posibilidad de aumentar alpha ("pasos"). Observamos que si aumentamos este alpha, la Función de Costes alcanza valores muy cercanos a cero con 100 interacciones, menos de las que hemos visto en los anteriores gráficos. Esto nos puede dar a entender que si se aumenta el alpha, se necesitarán menos interacciones para minimizar la Función de Costes.
